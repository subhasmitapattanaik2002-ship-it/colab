```python

for char in "Python":
    print(char)

# Iterating over a list
scores = [85, 92, 78, 95, 60]
total = 0
for score in scores:
    total += score
print("Total:", total)

# Iterating over a dictionary — keys by default
student = {"name": "Ravi", "age": 21, "grade": "A"}

for key in student:
    print(key)

for key, value in student.items():
    print(f"{key}: {value}")

for value in student.values():
    print(value)
```
